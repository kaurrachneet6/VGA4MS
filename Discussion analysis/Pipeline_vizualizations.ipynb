## Gait Video Study
Some vizualizations for the pipeline figure in the paper

```python
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from mpl_toolkits import mplot3d
```

```python
front3D_orig = pd.read_csv('C:\\Users\\Rachneet Kaur\\Box\\Gait Video Project\\figures\\frameworks_pic\\659.csv')
side3D_orig = pd.read_csv('C:\\Users\\Rachneet Kaur\\Box\\Gait Video Project\\figures\\frameworks_pic\\509.csv')
```

```python
front3D = front3D_orig.set_index(['Unnamed: 0']).reindex(['left toe 2', 'left heel', 'left toe 1', 'left heel', 'left ankle', \
                 'left knee', 'left hip', 'right hip', 'right knee', 'right ankle', 'right heel', 'right toe 1', \
                'right heel', 'right toe 2'])
front3D.reset_index(inplace=True)
```

```python
fig = plt.figure(figsize = (8, 8))
font = {'size'   : 12}
mpl.rc('font', **font)
ax = plt.axes(projection='3d')
ax.plot3D(front3D['y'], front3D['x'], front3D['z'], '-',  marker = 'o', color = 'tomato', markersize = 9, markerfacecolor = 'mediumspringgreen', markeredgecolor = 'k', linewidth = 3)
# for i in range(len(front3D_orig)): #plot each point + it's index as text above
#     ax.text(front3D_orig['y'][i],front3D_orig['x'][i],front3D_orig['z'][i],  '%s' % front3D_orig['Unnamed: 0'][i], size=10, zorder=1,  
#     color='k') 

ax.set_xlabel('y (treadmill length)', fontsize = 16)
ax.set_ylabel('x (treadmill width)', fontsize = 16)
ax.set_zlabel('z (treadmill height)', fontsize = 16)
plt.ylim([0, 87])
plt.xlim([0, 310])

plt.show()
```

```python
side3D = side3D_orig.set_index(['Unnamed: 0']).reindex(['left toe 2', 'left heel', 'left toe 1', 'left heel', 'left ankle', \
                 'right ankle', 'right heel', 'right toe 1', 'right heel', 'right toe 2'])
side3D.reset_index(inplace=True)
side3D = side3D[side3D['confidence']!=0]
```

```python
fig = plt.figure(figsize = (8, 8))
font = {'size'   : 12}
mpl.rc('font', **font)
ax = plt.axes(projection='3d')
ax.plot3D(side3D['y'], side3D['x'], side3D['z'], '-',  marker = 'o', color = 'tomato', markersize = 9, markerfacecolor = 'mediumspringgreen', markeredgecolor = 'k', linewidth = 3)
for i in range(len(side3D_orig)): #plot each point + it's index as text above
    ax.text(side3D_orig['y'][i],side3D_orig['x'][i],side3D_orig['z'][i],  '%s' % side3D_orig['Unnamed: 0'][i], size=10, zorder=1,  
    color='k') 

ax.set_xlabel('y (treadmill length)', fontsize = 16)
ax.set_ylabel('x (treadmill width)', fontsize = 16)
ax.set_zlabel('z (treadmill height)', fontsize = 16)
plt.ylim([0, 87])
plt.xlim([0, 310])
ax.set_zlim(10,120)
plt.show()
```
